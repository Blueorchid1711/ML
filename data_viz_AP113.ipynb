```python
data = pd.read_csv('final_vg.csv')
data.head()
```

```python
# Perform univariate analysis on numerical columns
numerical_cols = data.select_dtypes(include=np.number).columns

for col in numerical_cols:
    print(f"--- Analysis of {col} ---")
    print(data[col].describe())
    print("\n")
```

```python
# Perform univariate analysis on categorical columns
categorical_cols = data.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"--- Analysis of {col} ---")
    print(data[col].value_counts())
    print("\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar plot of the counts of each genre
plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='Genre', order=data['Genre'].value_counts().index, palette='viridis', hue='Genre', legend=False)
plt.title('Distribution of Video Game Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the value counts for the 'Genre' column
genre_counts = data['Genre'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 10))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Video Game Genres')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = data.select_dtypes(include=np.number).columns

for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(data=data, x=col, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = data.select_dtypes(include=np.number).columns

for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=data, x=col)
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```
